```python
import numpy as np
import tensorflow as tf
```

```python
mass = tf.Variable(np.load('q2_input/masses.npy'))
velocity = tf.Variable(np.load('q2_input/velocities.npy'))
position = np.load('q2_input/positions.npy')
epsilon =  tf.cast(tf.constant(10**(-4)), tf.float64)
pos = position.T
# print pos.shape
pos_rep = np.tile(pos, pos.shape[-1])
pos_rep = pos_rep.reshape(2,pos.shape[-1],pos.shape[-1])
pos_rep_t = np.transpose(pos_rep,(0,2,1))
r = pos_rep_t-pos_rep
G =  tf.cast(tf.constant(6.67*(10**(5))), tf.float64)
time = tf.placeholder(tf.float64)
r = tf.convert_to_tensor(r)
position = tf.convert_to_tensor(position)
velx = velocity[:,0]
vely = velocity[:,1]
posx = position[:,0]
posy = position[:,1]
```

```python
rd = -1*tf.math.sign(r)

#Add small value to prevent NaN on inverse
r += epsilon

# r ---> -(1/r^2)
r2 = rd * tf.math.reciprocal(tf.math.square(r))
f = G * tf.multiply(r2,mass)

# Get net acceleration on each particle
a = tf.reduce_sum(f, 2)
a = tf.transpose(a)
# Newton's Equations
new_pos = position + velocity*time + 0.5*(time**2)*a
new_vel = velocity + a*time
```

```python
init_g = tf.global_variables_initializer()
init_l = tf.local_variables_initializer()
with tf.Session() as sess:
    sess.run(init_g)
    sess.run(init_l)
#     print(sess.run(a))
    print(sess.run(new_pos, feed_dict={time: 0.001}))
    # Need to update with new value
```
